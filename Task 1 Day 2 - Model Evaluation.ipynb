## load data and define target and input

```python
#step 1 load data and define target and input
import pandas as pd
data = pd.read_csv("Downloads/CDSS 2025/Day 2/HR-Attrition.csv")

data.head()
```

```python
data.info()
```

```python
X = data.drop(columns=['Attrition']) 
y = data.Attrition #target 
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,  random_state = 0)
```

```python
#model evaluation using KNN (k-nearest neighbors)
from sklearn import neighbors
knn = neighbors.KNeighborsClassifier()
```

```python

knn.fit(X_train,y_train) #fitting the data into knn algorithm, and it will predict the logic behind input and target
```

```python

knn.score(X_test,y_test)
```

```python
#SVM (support vector machine)- classification
from sklearn import svm
svm = neighbors.KNeighborsClassifier()
```

```python
#SVM (support vector machine)- classification
from sklearn import svm
svm = svm.SVC()
```

```python
svm.fit(X_train,y_train) #fitting the data into knn algorithm, and it will predict the logic behind input and target
```

```python
svm.score(X_test,y_test)
```

```python
#Linear Regression - regression
from sklearn import linear_model
lm = linear_model.SGDClassifier()
```

```python
lm.fit(X_train,y_train) #fitting the data into knn algorithm, and it will predict the logic behind input and target
```

```python
lm.score(X_test,y_test)
```

```python
#Decision tree - classification
from sklearn import tree
dt = tree.DecisionTreeClassifier()
```

```python
dt.fit(X_train,y_train) #fitting the data into knn algorithm, and it will predict the logic behind input and target
```

```python
dt.score(X_test,y_test)
```
